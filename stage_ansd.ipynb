```python
import os
import rasterio
from rasterio.plot import show
import xml.etree.ElementTree as ET

```

```python
# Fonction pour traiter les fichiers JP2 et XML dans un rpertoire .SAFE
def traiter_safe(dossier_safe):
    for root, dirs, files in os.walk(dossier_safe):
        for file in files:
            file_path = os.path.join(root, file)
            if file.endswith('.jp2'):
                # Charger le fichier JP2 avec rasterio
                try:
                    with rasterio.open(file_path) as dataset:
                        # Afficher les informations sur le fichier
                        print(f"Informations sur {file}:")
                        print(dataset.profile)
                        # Afficher l'image
                        show(dataset)
                except Exception as e:
                    print(f"Erreur lors du traitement de {file}: {str(e)}")
            elif file.endswith('.xml'):
                # Charger et traiter le fichier XML
                try:
                    tree = ET.parse(file_path)
                    xml_root = tree.getroot()
                    # Afficher quelques mtadonnes
                    print(f"Metadonnes pour {file}:")
                    for child in xml_root:
                        print(child.tag, child.attrib)
                except Exception as e:
                    print(f"Erreur lors du traitement de {file}: {str(e)}")

# Chemin vers le rpertoire .SAFE
dossier_safe = r'C:\Users\ADIE\Documents\stage_ansd\images_satellite\S2A_MSIL1C_20240630T113321_N0510_R080_T28PDC_20240630T165124.SAFE'

# Appel de la fonction pour traiter le rpertoire .SAFE
traiter_safe(dossier_safe)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.windows import Window

# Fonction pour calculer NDVI
def calculate_ndvi(nir_band, red_band):
    ndvi = (nir_band - red_band) / (nir_band + red_band)
    return ndvi

# Fonction pour charger une partie des bandes et calculer NDVI
def load_and_preprocess_subset(file_path_b4, file_path_b8, window):
    with rasterio.open(file_path_b4) as red:
        red_band = red.read(1, window=window).astype(float)
    with rasterio.open(file_path_b8) as nir:
        nir_band = nir.read(1, window=window).astype(float)
    
    ndvi = calculate_ndvi(nir_band, red_band)
    return ndvi

# Chemin vers les fichiers JP2 pour les bandes rouge et NIR
file_path_b4 = r'C:\Users\ADIE\Documents\stage_ansd\images_satellite\S2A_MSIL1C_20240630T113321_N0510_R080_T28QDD_20240630T165124.SAFE\GRANULE\L1C_T28QDD_A047124_20240630T114240\IMG_DATA\T28QDD_20240630T113321_B04.jp2'  # Bande rouge
file_path_b8 = r'C:\Users\ADIE\Documents\stage_ansd\images_satellite\S2A_MSIL1C_20240630T113321_N0510_R080_T28QDD_20240630T165124.SAFE\GRANULE\L1C_T28QDD_A047124_20240630T114240\IMG_DATA\T28QDD_20240630T113321_B08.jp2'  # Bande NIR

# Dfinir une fentre d'intrt (bounding box)
col_off = 500  # Offset de la colonne de dpart
row_off = 7000  # Offset de la ligne de dpart
width = 2000    # Largeur de la fentre
height = 2000   # Hauteur de la fentre
window = Window(col_off, row_off, width, height)

# Charger et calculer NDVI pour la zone d'intrt
ndvi_subset = load_and_preprocess_subset(file_path_b4, file_path_b8, window)

# Afficher NDVI pour la zone d'intrt
plt.imshow(ndvi_subset, cmap='RdYlGn')
plt.colorbar()
plt.title('NDVI pour la zone d\'intrt')
plt.show()
```
